```python
import matplotlib.pyplot as plt
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```python
plt.scatter(df['beta'], df['RMSE'], marker='o')
plt.xscale('log')
plt.xlabel('Beta')
plt.ylabel('RMSE')
plt.title('Plot of RMSE vs. Beta')
plt.grid(True)
plt.show()
```

```python
plt.scatter(df['gamma'], df['RMSE'], marker='o')
plt.xscale('log')
plt.xlabel('Gamma')
plt.ylabel('RMSE')
plt.title('Plot of RMSE vs. Gamma')
plt.grid(True)
plt.show()
```

```python
plt.scatter(df['u'], df['RMSE'], marker='o')
plt.xscale('log')
plt.xlabel('u')
plt.ylabel('RMSE')
plt.title('Plot of RMSE vs. u')
plt.grid(True)
plt.show()
```

```python
plt.scatter(df['v'], df['RMSE'], marker='o')
plt.xscale('log')
plt.xlabel('v')
plt.ylabel('RMSE')
plt.title('Plot of RMSE vs. v')
plt.grid(True)
plt.show()
```

```python
import matplotlib.ticker as mticker
# Fixed values for gamma, u, and v
fixed_gamma = 0.001
fixed_u = 0.001
fixed_v = 0.001

# Filter the DataFrame based on fixed values
# Substitute values of 'alpha' equal to 0.001 with -1000
df['alpha'] = df['alpha'].replace(0.001, -1000)
# Substitute values of 'alpha' equal to 0.001 with -1000
df['beta'] = df['beta'].replace(0.001, -1000)
filtered_df = df[(df['gamma'] == fixed_gamma) & (df['u'] == fixed_u) & (df['v'] == fixed_v)]
print(filtered_df)
# Plotting
fig = plt.figure(figsize=(10, 6))  # Adjust the figsize as desired
ax = fig.add_subplot(111, projection='3d')

# Plotting the data points
ax.scatter(filtered_df['alpha'], filtered_df['beta'], filtered_df['RMSE'], c='b', marker='o')

ax.set_xlabel('Alpha')
ax.set_ylabel('Beta')
ax.set_zlabel('RMSE')
ax.set_title(f'RMSE vs. Alpha and Beta\nGamma={fixed_gamma}, U={fixed_u}, V={fixed_v}')

def log_tick_formatter(val, pos=None):
    if (val == -1000):
        return f"0.001"
    else:
        return val

# Set equally spaced ticks on x and y axes
# Set equally spaced ticks on x and y axes
ax.set_xticks([-1000, 1, 1000])
ax.set_yticks([-1000, 1, 1000])
ax.xaxis.set_major_formatter(mticker.FuncFormatter(log_tick_formatter))
ax.yaxis.set_major_formatter(mticker.FuncFormatter(log_tick_formatter))
# Add labels to the points
for x, y, z, label in zip(filtered_df['alpha'], filtered_df['beta'], filtered_df['RMSE'], filtered_df['RMSE']):
    ax.text(x, y, z, f'{label:.4f}', color='red')

# Display the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Fixed values for gamma, u, and v
fixed_gamma = 1.000
fixed_u = 1.000
fixed_v = 1.000

# Filter the DataFrame based on fixed values
# Substitute values of 'alpha' equal to 0.001 with -1000
df['alpha'] = df['alpha'].replace(-1000, 0.001)
# Substitute values of 'alpha' equal to 0.001 with -1000
df['beta'] = df['beta'].replace(-1000, 0.001)
filtered_df = df[(df['gamma'] == fixed_gamma) & (df['u'] == fixed_u) & (df['v'] == fixed_v)]

# Create a pivot table for heatmap
heatmap_data = filtered_df.pivot('beta', 'alpha', 'PSNR')

# Plotting the heatmap
plt.figure(figsize=(10, 6))  # Adjust the figsize as desired
sns.heatmap(heatmap_data, cmap='viridis', annot=True)

plt.xlabel('Alpha')
plt.ylabel('Beta')
plt.title(f'PSNR Heatmap\nGamma={fixed_gamma}, U={fixed_u}, V={fixed_v}')

# Display the plot
plt.show()
```
